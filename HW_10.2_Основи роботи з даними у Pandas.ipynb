### Hypothesis 1: Higher average unit price in the city
*   Rationale: if the product is sold at a higher price, while the costs remain similar, gross income will be higher
*   Testing: suggestion to compare average unit price across the cities, evaluate correlation between unit price and gross income
*   Additional data: income level of the population
*   Results: applying city specific pricing strategy; if the demand is not reduced then suggestion to test price increase in other cities

```python
avg_price_per_city = df.groupby ('City')['Unit price'].mean().sort_values(ascending=False)
print('Average unit price per city:')
print(avg_price_per_city)
```

```python
avg_gross_income_per_city = df.groupby('City')['gross income'].mean().sort_values(ascending=False)
print('Average gross income per city:')
print(avg_gross_income_per_city)
```

```python
correlation = df['Unit price'].corr(df['gross income'])
print('Correlation between unit price and gross income:', correlation)
```

There is a moderate positive correlation, higher unit price - higher income

### Hypothesis 2: Customers buy several items in one purchase

*   Rationale: if customers buy more units per transaction, the gross income also increases
*   Testing: compare average quantity per transaction across cities, analyze customer types as well as impact on purchase
*   Additional data: visitors count, demographic indicators
*   Results: where lower number of units identified per transaction, introduce promotions; encourage multi-item purchases





```python
avg_quantity_per_city = df.groupby('City')['Quantity'].mean().sort_values(ascending=False)
print('Average quantity per city :')
print(avg_quantity_per_city)
```

```python
avg_total_per_city = df.groupby('City')['Total'].mean().sort_values(ascending=False)
print('Average total per city:')
print(avg_total_per_city)
```

```python
correlation = df['Quantity'].corr(df['gross income'])
print('Overall correlation between quantity and gross income:', correlation)
```

There is a moderate positive correlation, higher quantity - higher income

```python
unique_invoices = df['Invoice ID'].nunique()
print('Number of unique invoices is', unique_invoices)
```

### Hypothesis 3: The city has a better (more profitable) product line
*   Rationale: if customers buy more expensive (high-margin products), the gross income will be higher
*   Testing: comparing the share of each product line across cities
*   Additional data: information on promotions, marketing campaigns, product lines data
*   Results: promote profitable product lines, optimize assortment accross stores, add high-margin products





```python
total_gross_income = df.groupby(['City', 'Product line'])['gross income'].sum()
total_gross_income_sorted = total_gross_income.groupby(level=0, group_keys= False).apply(lambda x:x.sort_values(ascending=False))
print(total_gross_income_sorted)
```

```python
gross_income_share = total_gross_income.groupby(level=0, group_keys=False).apply(lambda x: 100* x/x.sum()).round(2)
gross_income_share_sorted = gross_income_share.groupby(level=0,group_keys=False).apply(lambda x: x.sort_values())
print(gross_income_share_sorted)
```

9. Врахуовуючи висновок в попередньому завданні, дослідимо як продаються в цих містах різні категорії товарів. Виведіть відсотковий розпоіділ кількості інвойсів по категоріям товарів `Product line` в містах `Yangon` та `Mandalay`.
