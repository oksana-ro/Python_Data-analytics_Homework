```python
# Hypothesis 2: Customer type vs Product line (do members purchase more products than non-members?)
customer_type_products = df.groupby(['Customer type', 'Product line'])['Invoice ID'].count()
customer_type_products
```

### Analysis:

*   Normal customers buy more electronics and fashion accessoires
*   Members buy more home and lifestyle as well as sports and travel products
*   Health and beauty products are pretty similar for both customer types

### Recommendation:

*   For members: focus on food and beverages - introducing meal deals, points etc.
*   Use targgeted campaigns to convert normal customers to members
*   Increase cross-selling (e.g. home items & food and beverages)

```python
# Hypothesis 3: Time vs Product line (does hour affect the purchases of specific products?)
def get_hour(time_str):
    return int(time_str.split(':')[0])
df['Hour'] = df['Time'].apply(get_hour)
```

```python
hour_products = df.groupby(['Hour', 'Product line'])['Invoice ID'].count()
hour_products
```

```python
hour_product_line = df.pivot_table(index='Hour', columns='Product line', values='Invoice ID', aggfunc='count')
hour_product_line
```

```python
hour_product_line.style.highlight_max(axis=0)
```

### Analysis:
*   Electronics, food and beverages, fashion accessoires have a peak in the evening
*   Sports and travel ites - afternoon and evening time
*   Health and beauty items - afternoon time
*   Home & lifestyle items - morning and early afternoon

### Recommendation:
*   Use time-targeted promotions (evening and/or lunch time depending on product line)
*   Use personalized marketing
*   Increase the number of staff for processing orders in the evening (peak period)
